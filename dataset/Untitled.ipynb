```python
import pandas as pd
```

```python
df = pd.read_csv('data_cts_prisons_and_prisoners.csv')
```

```python
filtered_df = df[df['Unit of measurement'] == 'Counts']
filtered_df.head()
```

```python
grouped_df = filtered_df.groupby(['Region', 'Sex', 'Year'])['VALUE'].sum().reset_index()
grouped_df.head()
```

```python
grouped_df.columns = ['Region', 'Sex', 'Year', 'Total']
```

```python
grouped_df.head()
```

```python
grouped_df.to_csv('clean.csv', index=False)
```
